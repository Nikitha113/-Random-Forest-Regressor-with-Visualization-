```python
#RandomForestRegressor
#A Random Forest Regressor is a machine learning algorithm that predicts continuous values (numbers) by using many decision trees and taking the average of their predictions.
```

```python
#It builds many decision trees.
#Each tree gives its own prediction.
#The final output is the average of all tree outputs.
#This makes predictions more accurate and stable.
#Many trees together give a better answer than one tree.
```

```python
#Example
#Problem: Predict house price based on size.
#Tree 1 predicts: ₹50 lakhs
#Tree 2 predicts: ₹55 lakhs
#Tree 3 predicts: ₹52 lakhs
#Final Prediction = (50 + 55 + 52) / 3 = ₹52.33 lakhs
```

```python
import pandas as pd
import numpy as np

from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
# Example dataset: Study Hours vs Marks
data = {
    "Hours_Studied": [1, 2, 3, 4, 5, 6, 7, 8],
    "Marks":        [35, 40, 50, 55, 65, 70, 78, 85]
}
```

```python
df = pd.DataFrame(data)
print(df)
```

```python
X = df[["Hours_Studied"]]  # input
y = df["Marks"]           # output
```

```python
X = df[["Hours_Studied"]]  # input
y = df["Marks"]           # output
```

```python
#Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=0
)
```

```python
#Create Random Forest Regressor model
model = RandomForestRegressor(
    n_estimators=100,
    random_state=0
)
```

```python
#Train the model
model.fit(X_train, y_train)
```

```python
#Make predictions
y_pred = model.predict(X_test)
print("Predicted Marks:", y_pred)
```

```python
#Evaluate the model
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```

```python
#Predict for new input
new_hours = [[6]]
prediction = model.predict(new_hours)
print("Predicted marks for 6 study hours:", prediction)
```

```python
#Split X and y
X = df[["Hours_Studied"]]
y = df["Marks"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=0
)
```

```python
#Train Random Forest Regressor
model = RandomForestRegressor(n_estimators=10, random_state=0)
model.fit(X_train, y_train)
```

```python
#Predictions
y_pred = model.predict(X_test)
```

```python
#Matplotlib graph (Actual vs Predicted)
import matplotlib.pyplot as plt

plt.figure()
plt.scatter(X_test, y_test)
plt.scatter(X_test, y_pred)
plt.xlabel("Hours Studied")
plt.ylabel("Marks")
plt.title("Actual vs Predicted Marks")
plt.show()
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
plt.figure(figsize=(20,10))

plot_tree(
    model.estimators_[0],
    feature_names=["Hours_Studied"],
    filled=True,
    rounded=True
)

plt.title("Decision Tree from Random Forest")
plt.show()
```
